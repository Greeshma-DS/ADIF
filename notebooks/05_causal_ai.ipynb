```python
import pandas as pd

df = pd.read_csv(
    "../data/online_retail/clean_retail.csv",
    low_memory=False
)

df["InvoiceDate"] = pd.to_datetime(
    df["InvoiceDate"]
)

print(df.shape)
```

```python
monthly = (
    df.groupby(
        pd.Grouper(
            key="InvoiceDate",
            freq="ME"
        )
    )
    .agg({
        "Revenue":"sum",
        "Customer ID":"nunique",
        "Invoice":"nunique",
        "Quantity":"sum"
    })
)

monthly.columns = [
    "Revenue",
    "Customers",
    "Orders",
    "Quantity"
]

monthly = monthly.reset_index()

monthly.head()
```

```python
monthly["Revenue_Change"] = (
    monthly["Revenue"]
    .pct_change()
)

monthly["Customer_Change"] = (
    monthly["Customers"]
    .pct_change()
)

monthly["Order_Change"] = (
    monthly["Orders"]
    .pct_change()
)

monthly["Quantity_Change"] = (
    monthly["Quantity"]
    .pct_change()
)

monthly = monthly.dropna()

monthly.head()
```
