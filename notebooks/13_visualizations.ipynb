```python
import pandas as pd
import matplotlib.pyplot as plt

importance = pd.DataFrame({
    "Feature": [
        "Quantity Change",
        "Order Change"
    ],
    "Importance": [
        83.62,
        16.38
    ]
})

plt.figure(figsize=(8,5))

plt.bar(
    importance["Feature"],
    importance["Importance"]
)

plt.title(
    "Online Retail Root Cause Importance"
)

plt.ylabel("Importance (%)")

plt.savefig(
    "../results/online_retail_importance.png",
    bbox_inches="tight"
)

plt.show()
```
