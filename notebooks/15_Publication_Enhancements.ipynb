```python
from ollama import chat

prompt = f"""
You are a retail strategy consultant.

Root Cause Results:
{importance.to_string()}

Causal Effect:
ATE = {estimate.value}

Counterfactual Results:
{sensitivity.to_string()}

Generate executive recommendations.
"""

response = chat(
    model="llama3.1:8b",
    messages=[
        {
            "role":"user",
            "content":prompt
        }
    ]
)

report = response["message"]["content"]

with open(
    "../results/llm_generated_report.txt",
    "w",
    encoding="utf-8"
) as f:
    f.write(report)

print(report)
```

```python
import matplotlib.pyplot as plt

with open(
    "../results/llm_generated_report.txt",
    "r",
    encoding="utf-8"
) as f:
    text = f.read()

plt.figure(
    figsize=(10,8)
)

plt.axis("off")

plt.text(
    0,
    1,
    text,
    fontsize=10,
    wrap=True,
    va="top"
)

plt.savefig(
    "../results/figures/llm_recommendation_output.png",
    dpi=300,
    bbox_inches="tight"
)

plt.close()

print("Saved")
```
