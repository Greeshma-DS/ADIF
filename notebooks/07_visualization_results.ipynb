```python
import pandas as pd
import matplotlib
matplotlib.use("module://ipykernel.pylab.backend_inline", force=True)
import matplotlib.pyplot as plt

df = pd.read_csv(
    "../data/online_retail/clean_retail.csv",
    low_memory=False
)

df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

monthly_revenue = (
    df.groupby(
        pd.Grouper(
            key="InvoiceDate",
            freq="ME"
        )
    )["Revenue"]
    .sum()
)

plt.figure(figsize=(10,5))
plt.plot(monthly_revenue)
plt.title("Monthly Revenue Trend")
plt.show()
```
