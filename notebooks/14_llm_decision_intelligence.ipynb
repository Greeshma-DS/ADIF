```python
ate_value = 0.010647
```

```python
prompt = f"""
You are an enterprise decision intelligence assistant.

ONLY use the facts provided below.

Do NOT invent numbers.
Do NOT estimate revenue values.
Do NOT make assumptions.

FACTS

Root Cause Analysis:

{importance.to_string(index=False)}

Average Treatment Effect:
{ate_value}

Counterfactual Analysis:

{sensitivity.to_string(index=False)}

TASK

Generate:

1. Executive Summary
2. Key Findings
3. Strategic Recommendations

Every recommendation must explicitly reference the evidence above.

If information is unavailable, state:
'Not available from current analysis.'
"""
```

```python
from ollama import chat

response = chat(
    model="llama3.1:8b",
    messages=[
        {
            "role": "user",
            "content": prompt
        }
    ]
)

report = response["message"]["content"]

print(report)
```
